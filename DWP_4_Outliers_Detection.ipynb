```python
def detect_outlier_iqr(data_1):
  Q1=data_1.quantile(0.25)
  Q3=data_1.quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outlier=((data_1<lower_bound)|(data_1>upper_bound))
  return outlier

outlier_mask=data_1.apply(detect_outlier_iqr)
outlier_mask

```

```python
for column in data_1.columns:
  median_value=data_1[column].median
  data_1.loc[outlier_mask[column],column]=median_value
  print(column,median_value)
```
