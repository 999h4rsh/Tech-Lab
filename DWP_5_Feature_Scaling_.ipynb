Feature Scaling

Normalization: It is a scaling technique in which values are shifted and rescaled ranging between 0 and 1. Also known as mean max scalar.
Formula: x'= (x-Xmin)/(Xmax-Xmin)

Standardization: It is another scaling technique where the values are centered around the mean with a unique standard deviation. It follows the emiperical formula
Formula: z=(x-(miu))/s.d

If our data is not normally distributed then we use normalization.
If our data is normally distributed we use standard scaling method.

```python
import pandas as  pd
data=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv")
data
```

```python
#MinMax scalar
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scalar = MinMaxScaler()
normalize_data=scalar.fit_transform(data)
normalize_data
```

```python
#We name the columns
normalize_data=pd.DataFrame(normalize_data,columns=data.columns)
normalize_data
```

```python
#Find the max and min of each value
normalize_data.describe()
```

```python
#standard scalar on data
scalar=StandardScaler()
standardized_data=pd.DataFrame(scalar.fit_transform(data),columns=data.columns)
standardized_data
```

```python
standardized_data.describe()
```

```python
#Select Outliers
#Mean normalization
#x'=(x-min)/(xmax-xmin)
mean_norm=(data-data.mean())/(data.max()-data.min())
mean_norm
```

Maximum Absolute Scalar: used when data has 0 centereddistribution i.e most of the data lies near mean.

```python
#maximum_absolute_scaling
#x'=x/(max|x|)
from sklearn.preprocessing import MaxAbsScaler
scalar=MaxAbsScaler()
max_abs_scalar=pd.DataFrame(scalar.fit_transform(data),columns=data.columns)
max_abs_scalar
```
